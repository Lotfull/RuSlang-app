```python
import pandas as pd
teenslang_csv = '/Users/lotfull/Desktop/RuSlangApp/SlangDictionary/teenslang2/teenslang3.csv'
vsekidki_csv = '/Users/lotfull/Desktop/RuSlangApp/SlangDictionary/vsekidki2.csv'
modnyeslova_csv = '/Users/lotfull/Desktop/RuSlangApp/SlangDictionary/modnyeslova3.csv'
```

```python
teenslang = pd.read_csv(teenslang_csv)
vsekidki = pd.read_csv(vsekidki_csv)
modnyeslova = pd.read_csv(modnyeslova_csv)
```

```python
teenslang.head(2)
```

```python
teenslang.columns = ['name', 'definition', 'type', 'group', 'examples', 'origin', 'hashtags', 'synonyms', 'link']
teenslang.head(2)
```

```python
def a(cat):
    cat = eval(cat)
    if isinstance(cat, list):
        cat = ', '.join(cat)
    cat = cat.replace('Сложный прекол!', 'Сложный прикол')
    return cat

vsekidki.columns = ['link', 'definition', 'name', 'group']
vsekidki = vsekidki[['name', 'definition', 'group', 'link']]
vsekidki['group'] = vsekidki['group'].apply(a)
vsekidki['link'] = vsekidki['link'].apply(lambda link: 'vsekidki.ru/' + link)

vsekidki.head(2)
```

```python
modnyeslova.head(2)
```

```python
modnyeslova.columns = ['definition', 'link', 'name', 'group', 'video']
modnyeslova.head(2)
```

```python
teenslang['dict_id'] = 1
vsekidki['dict_id'] = 2
modnyeslova['dict_id'] = 3
```

```python
full = teenslang.append(vsekidki, sort=False).append(modnyeslova, sort=False)
```

```python
full.reset_index(inplace=True, drop=True)
```

```python
full.replace(' ', float('nan'))
```

```python
full.name = full.name.apply(lambda n: n[0].upper() + n[1:])
```

```python
full.to_csv('../SlangApp/full_dict.csv', index=False)
```
