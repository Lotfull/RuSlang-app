```python
import pandas as pd
teenslang_csv = '/Users/lotfull/Desktop/RuSlangApp/SlangDictionary/teenslang2/teenslang3.csv'
vsekidki_csv = '/Users/lotfull/Desktop/RuSlangApp/SlangDictionary/vsekidki3.csv'
modnyeslova_csv = '/Users/lotfull/Desktop/RuSlangApp/SlangDictionary/modnyeslova4.csv'
```

```python
teenslang = pd.read_csv(teenslang_csv)
vsekidki = pd.read_csv(vsekidki_csv)
modnyeslova = pd.read_csv(modnyeslova_csv)
```

```python
teenslang = teenslang[['name', 'definition', 'link', 'id', 'group', 'examples', 'hashtags', 'origin', 'synonyms', 'type']]
teenslang.head(2)
```

```python
vsekidki.head(2)
```

```python
modnyeslova['id'] = 70000 + modnyeslova.link.apply(lambda l: int(l.split('/')[1].split('-')[0]))
modnyeslova.columns = ['definition', 'link', 'name', 'group', 'video', 'id']
modnyeslova = modnyeslova[['name', 'definition', 'link', 'id', 'group', 'video']]
modnyeslova.head(2)
```

```python
teenslang['dict_id'] = 1
vsekidki['dict_id'] = 2
modnyeslova['dict_id'] = 3
```

```python
full = teenslang.append(vsekidki, sort=False).append(modnyeslova, sort=False)

full.reset_index(inplace=True, drop=True)

full.replace(' ', float('nan'))

full.name = full.name.apply(lambda n: n[0].upper() + n[1:])
```

```python
full.to_csv('../SlangApp/full_dict.csv', index=False)
print(full.columns)
full.head()
```
